```python
#Using iqr method to remove outliers for year and num_review cols
column='year'
q1= df[column].quantile(0.25)
q2= df[column].quantile(0.75)
IQR=q2-q1
lower_bound=q1-1.5*IQR
upper_bound=q2+1.5*IQR
df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
df_filtered.shape
```

```python
column='num_reviews'
q1= df[column].quantile(0.25)
q2= df[column].quantile(0.75)
IQR=q2-q1
lower_bound=q1-1.5*IQR
upper_bound=q2+1.5*IQR
df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
df_filtered.shape
```
