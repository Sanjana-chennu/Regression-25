```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("cleaned_wines_spafinal.csv")
```

```python
df.shape
```

```python
numerical_cols=df.select_dtypes(include=np.number).columns
numerical_cols=numerical_cols.drop('price')
for col in numerical_cols:
    sns.scatterplot(x=col,y='price',data=df)
    plt.title(f'Price vs.{col}')
    plt.xlabel(col)
    plt.ylabel('Price')
    plt.show()
```

```python
sns.scatterplot(x='body',y='price',data=df)
plt.title(f'Price vs. body')
plt.xlabel('body')
plt.ylabel('Price')
plt.show()
sns.scatterplot(x='acidity',y='price',data=df)
plt.title(f'Price vs. acidity')
plt.xlabel('acidity')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[numerical_cols].corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
numerical_cols=df.select_dtypes(include=np.number).columns
numerical_cols=numerical_cols.drop('price')
for col in numerical_cols:
    plt.bar(x=df[col],height=df['price'])
    plt.title(f'Price vs. {col}')
    plt.xlabel(col)
    plt.ylabel('Price')
    plt.show()

```

```python
#relationship bw body and acidity
plt.figure(figsize=(18,5))
plt.subplot(1,3,1)
sns.scatterplot(x=df['body'],y=df['acidity'],hue=df['price'],palette="coolwarm")
plt.title("Body vs Acidity (Color: Price)")

```

```python
# pie chart of years
import matplotlib.pyplot as plt
df['year'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribution of Years')
plt.ylabel('')
plt.show()
```

```python
#violin plot
for col in numerical_cols:
    sns.violinplot(x=col, y='price',data=df)
    plt.title(f'Price Distribution across {col}')
    plt.xlabel(col)
    plt.ylabel('Price')
    plt.show()
```
