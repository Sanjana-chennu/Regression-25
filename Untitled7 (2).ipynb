```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("cleaned_wines_spa.csv")
```

```python
df.shape
```

```python
sns.boxplot(y=df['year'])
```

```python
sns.boxplot(y=df['num_reviews'])
```

```python
#Using iqr method to remove outliers for year and num_review cols
column='year'
q1= df[column].quantile(0.25)
q2= df[column].quantile(0.75)
IQR=q2-q1
lower_bound=q1-1.5*IQR
upper_bound=q2+1.5*IQR
df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
df_filtered.shape
```

```python
column='num_reviews'
q1= df[column].quantile(0.25)
q2= df[column].quantile(0.75)
IQR=q2-q1
lower_bound=q1-1.5*IQR
upper_bound=q2+1.5*IQR
df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
df_filtered.shape
```

```python
df_filtered.to_csv("outliers.csv",index=False)
```

```python
#histplot to check skewness of numcols
num_cols=df_filtered.select_dtypes(include=['int64','float64']).columns
import matplotlib.pyplot as plt
for col in num_cols:
    sns.histplot(df_filtered[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_filtered['year_scaled']=scaler.fit_transform(df_filtered[['year']])
print(df_filtered['year_scaled'])
```

```python
from scipy import stats
# Applying Box-Cox transformation to 'num_reviews'
df_filtered['num_reviews_boxcox'], lambda_val = stats.boxcox(df_filtered['num_reviews'])
print(f"Lambda value for Box-Cox transformation: {lambda_val}")
sns.histplot(df_filtered['num_reviews_boxcox'], kde=True)
plt.title('Distribution of num_reviews after Box-Cox Transformation')
plt.show()
```

```python
#USE STANDARDIZATION FOR YEAR COL AND SHOWING HISTPLOT
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_filtered['year_standardized'] = scaler.fit_transform(df_filtered[['year']])
sns.histplot(df_filtered['year_standardized'],kde=True)
plt.title('Distribution of Year after Standardization')
plt.show()
```

```python
#Normalisation technique for num-rev col
import matplotlib.pyplot as plt
scaler=MinMaxScaler()
df_filtered['num_reviews_normalized']=scaler.fit_transform(df_filtered[['num_reviews']])
sns.histplot(df_filtered['num_reviews_normalized'],kde=True)
plt.title('Distribution of num_reviews after MinMax Normalization')
plt.show()
```

```python
#STANDARDIZATION FOR num_rev col
scaler = StandardScaler()
df_filtered['num_reviews_standardized'] = scaler.fit_transform(df_filtered[['num_reviews']])
sns.histplot(df_filtered['num_reviews_standardized'], kde=True)
plt.title('Distribution of num_reviews after Standardization')
plt.show()
```
